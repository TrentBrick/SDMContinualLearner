```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib import pylab as PP
import itertools

plt.style.use('ggplot') # change to ggplot style.
PP.rc('font', **{'sans-serif':'Arial', 'family':'sans-serif'}) # Uses Arial font for all.
```

### Find the best parameters for dense coding and 1L-NN

```python
dc_nn1 = pd.read_csv('./accuracy/dense-coding_1l-nn_result_lr0.1_minmax.csv',header=None)
dc_nn2 = pd.read_csv('./accuracy/dense-coding_1l-nn_result_lr0.01_minmax.csv',header=None)
dc_nn3 = pd.read_csv('./accuracy/dense-coding_1l-nn_result_lr0.001_minmax.csv',header=None)
dc_nn4 = pd.read_csv('./accuracy/dense-coding_1l-nn_result_lr0.0001_minmax.csv',header=None)
dc_nn5 = pd.read_csv('./accuracy/dense-coding_1l-nn_result_lr1e-05_minmax.csv',header=None)
```

```python
n = dc_nn1.shape[0]
dc_nn_1 = [np.mean(dc_nn1.iloc[i,:i+1]) for i in range(n)]
dc_nn_2 = [np.mean(dc_nn2.iloc[i,:i+1]) for i in range(n)]
dc_nn_3 = [np.mean(dc_nn3.iloc[i,:i+1]) for i in range(n)]
dc_nn_4 = [np.mean(dc_nn4.iloc[i,:i+1]) for i in range(n)]
dc_nn_5 = [np.mean(dc_nn5.iloc[i,:i+1]) for i in range(n)]
```

```python
plt.figure(figsize=(7,5))
x = np.array(range(1,11))*2
plt.plot(x, dc_nn_1,color='blue',label='1e-1',marker='.');
plt.plot(x, dc_nn_2,color='red',label='1e-2',marker='<');
plt.plot(x, dc_nn_3,color='black',label='1e-3',marker='s');
plt.plot(x, dc_nn_4,color='green',label='1e-4',marker='*');
plt.plot(x, dc_nn_5,color='green',label='1e-5',marker='+');
plt.legend(loc='best',fontsize=14);
```
