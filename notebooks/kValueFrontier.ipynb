```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
```

```python
# val accuracy and then final cont learning accuracy
kval_dict ={
    "1": ( .158,.737),
    "3": ( .143,.68), 
    "5": ( .129,.65),
    "8": ( .125,.626), 
    "10": ( .126,.61),
    "12": ( .129,.58), 
    "15": ( .137,.56),
    "18": ( .142,.536), 
    "20": ( .148,.50),
    "25": ( .16,.494), 
    "50": ( .206,.41),
    "100": ( .25,.376), 
    "150": ( .28,.355),
    "250": ( .3,.339), 
}
```

```python
k_vals = [k for k, v in kval_dict.items()]
all_val_accs = [v[0] for k, v in kval_dict.items()]
all_cont_accs = [v[1] for k, v in kval_dict.items()]
    
```

```python
plt.scatter(k_vals, all_val_accs, label="ImageNet32 Training")
plt.scatter(k_vals, all_cont_accs, label="Split CIFAR10 Cont Learning")
plt.legend()
plt.xlabel("K Value")
plt.ylabel("Validation Accuracy")
plt.title("Tradeoff between training and \n continual learning per k value")
plt.gcf().savefig('kValueAblate.png', dpi=250)
plt.show()
```

```python
plt.scatter(all_val_accs, all_cont_accs,  label="k Values")
plt.legend()
plt.xlabel("Validation Accuracy")
plt.ylabel("Cont Learning Accuracy")
plt.title("Pareto Frontier between training and continual learning")
plt.gcf().savefig('kValuePareto.png', dpi=250)
plt.show()
```
